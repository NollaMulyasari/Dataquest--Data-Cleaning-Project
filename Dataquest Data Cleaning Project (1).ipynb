from looking at these rows, we can see that DBN column in sat_results data is just the combination of CSD column and school code columns in the class_size. The difference is that DBN is padded, so the CSD portion of it alwys consist of two digits. That means we will need to add a leading 0 to the CSD if the CSD is less than two digits long

```python
# copy the dbn column in hs_directory into a new column called DBN

data['hs_directory']['DBN'] = data['hs_directory']['dbn']

# create a new column called padded_csd in the class_size data set

def pad_csd(num):
    return str(num).zfill(2)

data["class_size"]['padded_csd']= data['class_size']['CSD'].apply(pad_csd)
data["class_size"]["DBN"] = data['class_size']['padded_csd'] + data['class_size']['SCHOOL CODE']

data['class_size'].head()
```

we need to convert the SAT Math Avg. Score, SAT Critical Reading Avg.Score, and SAT Writing Avg. Score columns in the sat_results data set from the data type string to a numeric data type

```python
numeric_cols = ['SAT Math Avg. Score','SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']

for col in numeric_cols:
    data['sat_results'][col]= pd.to_numeric(data['sat_results'][col], errors= 'coerce')

# create a column called sat_score in sat_results that holds the combined SAT score for each student

data['sat_results']['sat_score']= data['sat_results'][numeric_cols[0]]+ data['sat_results'][numeric_cols[1]] + data['sat_results'][numeric_cols[2]]

data['sat_results']['sat_score'].head()
```

Next, let's parse the latitude and longitude coordibates for each school

```python
import re

def find_lat(loc):
    coords= re.findall("\(.+\)", loc)
    lat= coords[0].split(",")[0].replace("(", "")
    return lat
#assign the result to the lat column of hs_directory
data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(find_lat)

data["hs_directory"].head()
```

```python
def find_lon(loc):
    coords= re.findall("\(.+\)", loc)
    lon = coords[0].split(",")[1].replace(")", "").strip()
    return lon

data["hs_directory"]["lon"]= data["hs_directory"]["Location 1"].apply(find_lon)

data["hs_directory"]["lat"] = pd.to_numeric(data["hs_directory"]["lat"], errors="coerce")
data["hs_directory"]["lon"] = pd.to_numeric(data["hs_directory"]["lon"], errors="coerce")

data["hs_directory"].head()
```
